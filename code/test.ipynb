```python
import glob, os, csv, re, json, shutil
from PIL import Image
from openpyxl import Workbook
from openpyxl import load_workbook
from openpyxl import Workbook
from typing_extensions import final

main_path = 'Z:/학습데이터/20200612'
jpg_list = glob.glob('{}/*.jpg'.format(main_path))

for jpg_name in jpg_list:
    row_jpg = jpg_name.split('\\')[-1]
    number = re.findall('\d{5}',row_jpg)
    new_number= int(''.join(number)) + 100000
    json_name = re.sub('.jpg', '.json', jpg_name)
    new_jpg = re.sub('\d{5}', str(new_number), row_jpg)   # 반전된 jpg이름
    new_json = new_jpg.replace('.jpg', '.json')   # 반전된 json이름
    re_path = jpg_name.split('\\')[0]+'_re'
    
    image1 = Image.open(jpg_name)   #이미지 불러오기
    FlipImage = image1.transpose(Image.FLIP_LEFT_RIGHT)   #이미지 좌우대칭
    if not (os.path.isdir(re_path)):   # _re폴더 생성
        os.makedirs(re_path)    
    FlipImage.save(os.path.join(re_path, new_jpg))
    c_d_class = []
    try:
        with open(json_name, encoding='utf-8') as json_file:
            json_data = json.load(json_file)
            json_data['image']['filename'] = new_jpg
            annotation = json_data['annotations']
            
            
            for i in range(len(annotation)):
                if 'box' in annotation[i]:
                    annotation[i]['box'][0] = 1920 - annotation[i]['box'][0]
                    annotation[i]['box'][2] = 1920 - annotation[i]['box'][2]

                if not annotation[i]["middle classification"] in ['01', '02']:
                    c_d_class.append(annotation[i])
            
            class_c_d = {"image": json_data["image"], "annotations": c_d_class}
        
            if class_c_d["annotations"]:
                with open(os.path.join(re_path, new_json), 'w', encoding='utf-8') as json_change:
                    json.dump(class_c_d, json_change, indent=2, ensure_ascii=False)
            class_c_d.clear()  
    finally:
        json_file.close()

before_path = main_path+'_re'
after_path = main_path+'_re'
dir_path = main_path.replace(main_path.split('/')[-1], '')
os.chdir(main_path+'_re')
if not os.path.isdir('../카운트'):
    os.makedirs('../카운트')


write_xl = Workbook()
write_ws = write_xl.active
write_ws.append(['폴더명', 'JSON', 'class', 'class_count', 'box', 'polygon', 'point'])
for (path, dir, files) in os.walk(main_path+'_re'):
    firstRow = True
    for fileName in files:
        if fileName.endswith('.json'):
            firstJson = True
            with open(os.path.join(path, fileName), encoding='utf-8') as json_file:
                try:
                    jsonData = json.load(json_file)
                    annotations = jsonData['annotations']
                    classType = []
                    classTypeList = ['box', 'polygon', 'point']
                    classCountDict = {}
                    classTypeCountDict = {}
                    

                    for i in range(len(annotations)):
                        classType.append(annotations[i]['class'])

                    for key in classType:
                        classCountDict[key] = classType.count(key)
                        classTypeCountDict[key] = {'box': 0, 'polygon': 0, 'point': 0}

                    for key in classCountDict:
                        for i in range(len(annotations)):
                            if annotations[i]['class'] == key:
                                for item in classTypeList:
                                    if item in annotations[i]:
                                        classTypeCountDict[key][item] += 1

                    for key in classCountDict:
                        if firstRow and firstJson:
                            write_ws.append([path, fileName, key, classCountDict[key],
                                            classTypeCountDict[key][classTypeList[0]],
                                            classTypeCountDict[key][classTypeList[1]],
                                            classTypeCountDict[key][classTypeList[2]]])
                            firstRow = False
                            firstJson = False
                        elif firstJson:
                            write_ws.append(["", fileName, key, classCountDict[key],
                                            classTypeCountDict[key][classTypeList[0]],
                                            classTypeCountDict[key][classTypeList[1]],
                                            classTypeCountDict[key][classTypeList[2]]])
                            firstJson = False
                        else:
                            write_ws.append(["", "", key, classCountDict[key],
                                            classTypeCountDict[key][classTypeList[0]],
                                            classTypeCountDict[key][classTypeList[1]],
                                            classTypeCountDict[key][classTypeList[2]]])
                except Exception as ex:
                    print(ex)
                    pass

write_xl.save(os.path.join(main_path, '../카운트/{}_re.xlsx'.format(main_path.split('/')[-1])))
print(main_path + '완료')
```

```python
import glob, os, csv, re, json, shutil
from PIL import Image
from openpyxl import Workbook
from openpyxl import load_workbook
from openpyxl import Workbook
from typing_extensions import final
# main_path = 'D:/학습데이터/20201105'
c_d_class = []
main_path = 'F:/새 폴더/aa'
jpg_list = glob.glob('{}/*.jpg'.format(main_path))

for jpg_name in jpg_list:
    row_jpg = jpg_name.split('\\')[-1]
    number = re.findall('\d{5}',row_jpg)
    new_number= int(''.join(number)) + 100000
    json_name = re.sub('.jpg', '.json', jpg_name)
    new_jpg = re.sub('\d{5}', str(new_number), row_jpg)   # 반전된 jpg이름
    new_json = new_jpg.replace('.jpg', '.json')   # 반전된 json이름
    re_path = jpg_name.split('\\')[0]+'_re'
    
    image1 = Image.open(jpg_name)   #이미지 불러오기
    FlipImage = image1.transpose(Image.FLIP_LEFT_RIGHT)   #이미지 좌우대칭
    if not (os.path.isdir(re_path)):   # _re폴더 생성
        os.makedirs(re_path)    
    FlipImage.save(os.path.join(re_path, new_jpg))

    try:
        with open(json_name, encoding='utf-8') as json_file:
            json_data = json.load(json_file)
            json_data['image']['filename'] = new_jpg
            annotation = json_data['annotations']
            
            
            for i in range(len(annotation)):
                annotation[i]['box'][0] = 1920-annotation[i]['box'][0]
                annotation[i]['box'][2] = 1920-annotation[i]['box'][2]

                if not annotation[i]["middle classification"] in ['01', '02']:
                    c_d_class.append(annotation[i])
            
            class_c_d = {"image": json_data["image"], "annotations": c_d_class}
        
            if class_c_d["annotations"]:
                with open(os.path.join(re_path, new_json), 'w', encoding='utf-8') as json_change:
                    json.dump(class_c_d, json_change, indent=2, ensure_ascii=False)
            class_c_d.clear()  
    finally:
        json_file.close()

before_path = main_path+'_re'
after_path = main_path+'_re'
dir_path = main_path.replace(main_path.split('/')[-1], '')
os.chdir(main_path+'_re')
if not os.path.isdir('../카운트'):
    os.makedirs('../카운트')


write_xl = Workbook()
write_ws = write_xl.active
write_ws.append(['폴더명', 'JSON', 'class', 'class_count', 'box', 'polygon', 'point'])
for (path, dir, files) in os.walk(main_path+'_re'):
    firstRow = True
    for fileName in files:
        if fileName.endswith('.json'):
            firstJson = True
            with open(os.path.join(path, fileName), encoding='utf-8') as json_file:
                try:
                    jsonData = json.load(json_file)
                    annotations = jsonData['annotations']
                    classType = []
                    classTypeList = ['box', 'polygon', 'point']
                    classCountDict = {}
                    classTypeCountDict = {}
                    

                    for i in range(len(annotations)):
                        classType.append(annotations[i]['class'])

                    for key in classType:
                        classCountDict[key] = classType.count(key)
                        classTypeCountDict[key] = {'box': 0, 'polygon': 0, 'point': 0}

                    for key in classCountDict:
                        for i in range(len(annotations)):
                            if annotations[i]['class'] == key:
                                for item in classTypeList:
                                    if item in annotations[i]:
                                        classTypeCountDict[key][item] += 1

                    for key in classCountDict:
                        if firstRow and firstJson:
                            write_ws.append([path, fileName, key, classCountDict[key],
                                            classTypeCountDict[key][classTypeList[0]],
                                            classTypeCountDict[key][classTypeList[1]],
                                            classTypeCountDict[key][classTypeList[2]]])
                            firstRow = False
                            firstJson = False
                        elif firstJson:
                            write_ws.append(["", fileName, key, classCountDict[key],
                                            classTypeCountDict[key][classTypeList[0]],
                                            classTypeCountDict[key][classTypeList[1]],
                                            classTypeCountDict[key][classTypeList[2]]])
                            firstJson = False
                        else:
                            write_ws.append(["", "", key, classCountDict[key],
                                            classTypeCountDict[key][classTypeList[0]],
                                            classTypeCountDict[key][classTypeList[1]],
                                            classTypeCountDict[key][classTypeList[2]]])
                except Exception as ex:
                    print(ex)
                    pass

write_xl.save(os.path.join(main_path, '../카운트/{}_re.xlsx'.format(main_path.split('/')[-1])))
print(main_path + '완료')
```

```python
# -*- coding: UTF-8 -*-
import os, json

before_path = 'C:/Users/S2_SVR/Desktop/새 폴더/새 폴더 (3)/BCD만/20201030_S2-N2105M49064'
a_path = 'C:/Users/S2_SVR/Desktop/새 폴더/새 폴더 (3)/middle classification_01'
b_path = 'C:/Users/S2_SVR/Desktop/새 폴더/새 폴더 (3)/middle classification_02'
c_path = 'C:/Users/S2_SVR/Desktop/새 폴더/새 폴더 (3)/middle classification_03_06'
d_path = 'C:/Users/S2_SVR/Desktop/새 폴더/새 폴더 (3)/middle classification_07'

### 중분류
a = ['01']
b = ['02']
c = ['03','04','05','06']
d = ['07']

### 폴더 생성
if not (os.path.isdir(a_path)):
    os.makedirs(a_path)
if not (os.path.isdir(b_path)):
    os.makedirs(b_path)    
if not (os.path.isdir(c_path)):
    os.makedirs(c_path)
if not (os.path.isdir(d_path)):
    os.makedirs(d_path)
    
### 클래스 생성
a_class = []
b_class = []
c_class = []
d_class = []

for (path, dir, files) in os.walk(before_path):
    for fileName in files:
        if fileName.endswith('.json'):
            try:
                 with open(os.path.join(path, fileName),encoding='utf-8') as json_file:  # 파일 열기
                    jsonData = json.load(json_file)  # 디스크에 있는 포맷 데이터를 파이썬으로 읽어 오기
                    jsonimge = jsonData["image"]  # json이미지 key추출

                    for jsonanno in jsonData["annotations"]:
                        if jsonanno["middle classification"] in a:
                            a_class.append(jsonanno)
                        if jsonanno["middle classification"] in b:
                            b_class.append(jsonanno)
                        if jsonanno["middle classification"] in c:
                            c_class.append(jsonanno)
                        if jsonanno["middle classification"] in d:
                            d_class.append(jsonanno)

                    # 딕셔너리 생성
                    class_a_dic = {"image": jsonimge, "annotations": a_class}
                    class_b_dic = {"image": jsonimge, "annotations": b_class}
                    class_c_dic = {"image": jsonimge, "annotations": c_class}
                    class_d_dic = {"image": jsonimge, "annotations": d_class}

                    # 값이 존재할 경우 json데이터 저장
                    if class_a_dic["annotations"]:
                        with open(os.path.join(a_path, fileName), 'w', encoding='utf-8') as json_change:
                            json.dump(class_a_dic, json_change, indent=2, ensure_ascii=False)
                    if class_b_dic["annotations"]:
                        with open(os.path.join(b_path, fileName), 'w', encoding='utf-8') as json_change:
                            json.dump(class_b_dic, json_change, indent=2, ensure_ascii=False)
                    if class_c_dic["annotations"]:
                        with open(os.path.join(c_path, fileName), 'w', encoding='utf-8') as json_change:
                            json.dump(class_c_dic, json_change, indent=2, ensure_ascii=False)
                    if class_d_dic["annotations"]:
                        with open(os.path.join(d_path, fileName), 'w', encoding='utf-8') as json_change:
                            json.dump(class_d_dic, json_change, indent=2, ensure_ascii=False)

                    #  리스트 초기화
                    a_class.clear()
                    b_class.clear()
                    c_class.clear()
                    d_class.clear()

            finally:
                json_file.close()
```
