### Choose only location = [1920, 1440]

```python
new_list = []
for i in range(len(df)):
    if df['size'][i] == [1920, 1440]:
        new_list.append(df.iloc[i])
df_1920 = pd.DataFrame(new_list)
df_1920.reset_index(inplace=True)
df_1920.rename(columns={'index':'file_name'}, inplace=True)
df_1920.set_index('file_name', inplace = True)
df_1920.head(10)
```

### Make_data_set

```python
class_dict = {}

with open('/data/ai_dataSet/Yolov4/TensorFlow-2.x-YOLOv3/custom_set/classes.txt', encoding='utf-8') as aa :
    for i,j in enumerate(aa):
        class_dict[j.split('\n')[0]] = i 

for i in tqdm_notebook(df_1920.index, desc = 'Convert_to_data'):
    path = '/data/ai_dataSet/Yolov4/TensorFlow-2.x-YOLOv3/custom_set/ai_set/'
    try:
        with open(path + i.replace('.jpg','.json'), encoding='utf-8') as json_file:
            json_data = json.load(json_file)
            ann = json_data['regions'][0]
            class_id = str(class_dict[ann['class']])
            x1, y1, x2, y2 = str(ann['boxcorners'][0])+',', str(ann['boxcorners'][1])+',', str(ann['boxcorners'][2])+',', str(ann['boxcorners'][3])+','
            with open('/data/ai_dataSet/Yolov4/TensorFlow-2.x-YOLOv3/custom_set/1920_1440.txt', 'a', encoding='utf-8') as F:
                F.write(path+i + ' ' + x1 + y1 + x2 + y2 + class_id +'\n')
                F.close()
    except Exception as ex:
        print(ex)
        print(i)
```

### Split_train_test_validation

```python
X = []

with open('/data/ai_dataSet/Yolov4/TensorFlow-2.x-YOLOv3/custom_set/1920_1440.txt', encoding='utf-8') as f:
    for i in f:
        X.append(i.split('\n')[0])

train_ratio = 0.75
validation_ratio = 0.15
test_ratio = 0.10


# train is now 75% of the entire data set
# the _junk suffix means that we drop that variable completely
x_train, x_test = train_test_split(X, test_size=1 - train_ratio)

# test is now 10% of the initial data set
# validation is now 15% of the initial data set
x_val, x_test = train_test_split(x_test, test_size=test_ratio/(test_ratio + validation_ratio)) 

for i in tqdm_notebook(x_train, desc = 'train'):
    with open('/data/ai_dataSet/Yolov4/TensorFlow-2.x-YOLOv3/custom_set/train.txt','a', encoding='utf-8') as f:
            f.write(i+'\n')
            f.close()
for i in tqdm_notebook(x_val, desc = 'val'):
    with open('/data/ai_dataSet/Yolov4/TensorFlow-2.x-YOLOv3/custom_set/val.txt','a', encoding='utf-8') as f:
            f.write(i+'\n')
            f.close()
for i in tqdm_notebook(x_test, desc = 'test'):
    with open('/data/ai_dataSet/Yolov4/TensorFlow-2.x-YOLOv3/custom_set/test.txt','a', encoding='utf-8') as f:
            f.write(i+'\n')
            f.close()
```
