### 일간 변동률로 주가 비교

일간 변동률 = [(오늘종가 - 어제종가) / 어제종가] * 100

* 액면 분할이란?
    - 주식 분할은 시가 총액의 변화없이 기존 주식을 세분화하는 것이다. 일정한 비율로 액면가를 나눠 그 주식수를 증식시키는데서 액면 분할이라 함.

```python
# 삼성전자 일간 변동률
print(type(sec['Close']))
print('삼성전자 오늘의 종가 ')
print(sec['Close'])
print('-'*40)

# 어제의 종가 구하기 => 이전 거래일의 종가
print('삼성전자 어제의 종가 :')
print(sec['Close'].shift(1))
print('-'*40)

sec_dpc = (sec['Close']/sec['Close'].shift(1)-1)*100
print('삼성전자의 일간 변동률 :\n', sec_dpc.head())

print('-'*40)
print(type(msft['Close']))
print('마이크로소프트 오늘의 종가 ')
print(msft['Close'])
print('-'*40)

# 어제의 종가 구하기 => 이전 거래일의 종가
print('마이크로소프트 어제의 종가 :')
print(msft['Close'].shift(1))
print('-'*40)

msft_dpc = (msft['Close']/msft['Close'].shift(1)-1)*100
print('마이크로소프트 일간 변동률 :\n', msft_dpc.head())
```

### 주가 일간 변동률 시각화

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
import warnings
warnings.filterwarnings(action='ignore')

sec_dpc = (sec['Close']/sec['Close'].shift(1)-1)*100
sec_dpc.fillna(0)
msft_dpc = (msft['Close']/msft['Close'].shift(1)-1)*100
msft_dpc.fillna(0)

plt.hist(sec_dpc, bins=18)
plt.grid(True)
plt.title('Samsung Electronics Daily Percent Change')
plt.savefig('Daily Percent Change(Samsumg).png', bbox_inches='tight') # 그래프가 0을 기준으로 대칭적인 모습을 보임
plt.show()
# 기초 통계량 확인
display('Samsung Electronics :', sec_dpc.describe())

plt.hist(msft_dpc, bins=18)
plt.grid(True)
plt.title('Microsoft Daily Percent Change')

# 기초 통계량 확인
display('Microsoft :', sec_dpc.describe())
plt.savefig('Daily Percent Change(Microsoft).png', bbox_inches='tight')
plt.show()
```

```python
# 일간 변동률 누적합
# cumsum => series에서 제공하는 누적합 함수
sec_dpc_cs = sec_dpc.cumsum()
print('삼성전자 일간 변동 누적합 :\n', sec_dpc_cs)
print('-'*40)
msft_dpc_cs = msft_dpc.cumsum()
print('마이크로소프트 일간 변동 누적합 :\n', msft_dpc_cs)
```

### 삼성전자와 마이크로소프트 주식수익률 비교

```python
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()

# Samsung
sec = pdr.get_data_yahoo('005930.KS', start ='2019-10-01')
sec_dpc = ((sec['Close']-sec['Close'].shift(1)) / sec['Close'].shift(1)) * 100
sec_dpc.iloc[0] = 0 # 일간 변동률의 첫 행의 값인 NaN값 처리
sec_dpc_cs = sec_dpc.cumsum()

# Microsoft
msft = pdr.get_data_yahoo('MSFT', start ='2019-10-01')
msft_dpc = ((msft['Close']-msft['Close'].shift(1)) / msft['Close'].shift(1)) * 100
msft_dpc.iloc[0] = 0 # 일간 변동률의 첫 행의 값인 NaN값 처리
msft_dpc_cs = msft_dpc.cumsum()

import matplotlib.pyplot as plt
plt.plot(sec.index, sec_dpc_cs, 'b', label = 'Samsung Electronics')
plt.plot(msft.index, msft_dpc_cs, 'r--', label = 'Microsoft')
plt.ylabel('Change %')
plt.grid(True)
plt.title('Microsoft&Samsung Electronics(cumsum)')
plt.legend(loc = 'best')

# 이미지 저장
plt.savefig('Daily Percent Change(cumsum).png', bbox_inches='tight')
```

### 최대 손실 낙폭(MDD)

- MDD = (최저점 - 최고점) / 최저점
- KOSPI 지수 2500이란? 1980년 1월 4일에 상장된 모든 종목의 시가 총액을 기준으로 25배가 증가됨을 의미
- series.rolling(윈도우 크기 [,min_periods=1]) [.집계함수]
    - rolling 함수는 series에서 윈도우 크기에 해당하는 개수만큼 데이터를 추출하여 집계함수에 해당하는 연산을 실시.(max, min, mean, ...)
         - min_periods를 지정하면 데이터 개수가 윈도우 크기에 못미치더라도 min_periods로 지정한 개수만 만족하면 연산을 수행

```python
# KOSPI의 MDD 구하기
kospi = pdr.get_data_yahoo('^KS11',start = '2004-01-04') # KOSPI지수 데이터 다운로드
# sec = pdr.get_data_yahoo('005930.KS', start ='2019-10-01')

window = 252 # 산정기간에 해당하는 window값은 1년 동안의 개장일을 252일로 어림잡아 설정
peak = kospi['Adj Close'].rolling(window, min_periods=1).max() # KOSPI종가 컬럼에서 1년 기간단위로 MaxPeak
drawdown = kospi['Adj Close']/peak - 1.0 # MaxPeak대비 현재 KOSPI 종가가 얼마나 하락했는지
max_dd = drawdown.rolling(window, min_periods=1).min() # 1년 기간단위로 최저치 => 마이너스값이기 때문에 최저치가 바로 MDD

plt.figure(figsize=(20,18))
plt.subplot(211) # 2행 1열 중 1행
kospi['Close'].plot(label = "KOSPI", title = "KOSPI MDD", grid = True, legend = True)
plt.subplot(212)
drawdown.plot(c = 'b', label = "KOSPI DD", grid = True, legend = True)
max_dd.plot(c = 'r', label = "KOSPI MDD", grid = True, legend = True)

# plt.savefig('KOSPI_MDD.png', bbox_inches='tight')
```

```python
# 최대 낙폭 정확한 수치
print('최대 낙폭 :',max_dd.min())

# 최댁 낙폭 날짜
max_dd[max_dd==max_dd.min()] # 2008년 10월 24일 부터 2009년 10월 22일까지 252일(1년) 동안 주어진 max_dd가 일치
```
